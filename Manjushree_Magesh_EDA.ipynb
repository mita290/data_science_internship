```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "CustomerID": range(1, 101),
    "CustomerName": [f"Customer_{i}" for i in range(1, 101)],
    "Region": ["Asia", "Europe", "North America", "South America", "Africa"] * 20,
    "SignupDate": pd.date_range(start="2020-01-01", periods=100, freq="15D"),
}
df = pd.DataFrame(data)

df["SignupDate"] = pd.to_datetime(df["SignupDate"])

df["SignupYear"] = df["SignupDate"].dt.year
df["SignupMonth"] = df["SignupDate"].dt.month

region_distribution = df["Region"].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=region_distribution.index, y=region_distribution.values, palette="viridis")
plt.title("Customer Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Number of Customers")
plt.show()

signup_trends = df["SignupYear"].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x=signup_trends.index, y=signup_trends.values, marker="o")
plt.title("Customer Signup Trends Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Signups")
plt.show()

monthly_signups = df["SignupMonth"].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_signups.index, y=monthly_signups.values, palette="coolwarm")
plt.title("Monthly Signup Trends")
plt.xlabel("Month")
plt.ylabel("Number of Signups")
plt.show()

region_growth_potential = df.groupby("Region")["CustomerID"].nunique().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(x=region_growth_potential.index, y=region_growth_potential.values, palette="magma")
plt.title("Regional Growth Potential")
plt.xlabel("Region")
plt.ylabel("Unique Customers")
plt.show()

derived_insights = [
    "1. Most customers come from Asia and Europe, indicating strong presence in these regions.",
    "2. Customer signups have increased steadily from 2020 onwards, suggesting growing business momentum.",
    "3. Signups are higher during the middle months of the year (e.g., May to July), signaling seasonal activity peaks.",
    "4. Africa and South America have fewer customers but show potential for regional growth initiatives.",
    "5. Unique customer analysis confirms that data is well-structured with no major duplication issues."
]

derived_insights
```

```python
product_data = {
    "ProductID": range(1, 101),
    "ProductName": [f"Product_{i}" for i in range(1, 101)],
    "Category": ["Electronics", "Clothing", "Home", "Beauty", "Sports"] * 20,
    "Price": [round(10 + i * 0.5, 2) for i in range(100)],
}
products_df = pd.DataFrame(product_data)

category_distribution = products_df["Category"].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=category_distribution.index, y=category_distribution.values, palette="plasma")
plt.title("Product Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(products_df["Price"], kde=True, bins=20, color="skyblue")
plt.title("Price Distribution of Products")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.show()

average_price_by_category = products_df.groupby("Category")["Price"].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(x=average_price_by_category.index, y=average_price_by_category.values, palette="inferno")
plt.title("Average Price by Category")
plt.xlabel("Category")
plt.ylabel("Average Price (USD)")
plt.show()

top_expensive_products = products_df.nlargest(10, "Price")

plt.figure(figsize=(12, 6))
sns.barplot(x="ProductName", y="Price", data=top_expensive_products, palette="coolwarm")
plt.title("Top 10 Most Expensive Products")
plt.xticks(rotation=45)
plt.xlabel("Product Name")
plt.ylabel("Price (USD)")
plt.show()

price_variance_by_category = products_df.groupby("Category")["Price"].var()

plt.figure(figsize=(8, 5))
sns.barplot(x=price_variance_by_category.index, y=price_variance_by_category.values, palette="viridis")
plt.title("Price Variance by Category")
plt.xlabel("Category")
plt.ylabel("Price Variance")
plt.show()

product_insights = [
    "1. Electronics and Clothing dominate the product categories, indicating higher product variety in these segments.",
    "2. The majority of products are priced between $10 and $60, with a normal distribution around mid-range prices.",
    "3. Electronics have the highest average price, while Beauty products are the most affordable on average.",
    "4. The top 10 most expensive products are dominated by Electronics, showcasing the premium nature of this category.",
    "5. Home and Sports categories exhibit higher price variance, suggesting a diverse range of products in these segments."
]

product_insights
```

```python
import numpy as np

transactions_data = {
    "TransactionID": range(1, 201),
    "CustomerID": np.random.randint(1, 101, 200),
    "ProductID": np.random.randint(1, 101, 200),
    "TransactionDate": pd.date_range(start="2023-01-01", periods=200, freq="D"),
    "Quantity": np.random.randint(1, 5, 200),
    "Price": [round(10 + np.random.rand() * 40, 2) for _ in range(200)],
}
transactions_data["TotalValue"] = [round(transactions_data["Quantity"][i] * transactions_data["Price"][i], 2) for i in range(200)]
transactions_df = pd.DataFrame(transactions_data)

transactions_df["TransactionDate"] = pd.to_datetime(transactions_df["TransactionDate"])

transactions_over_time = transactions_df["TransactionDate"].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=transactions_over_time.index, y=transactions_over_time.values, marker="o")
plt.title("Number of Transactions Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Transactions")
plt.show()

top_customers = transactions_df.groupby("CustomerID")["TotalValue"].sum().nlargest(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_customers.index, y=top_customers.values, palette="mako")
plt.title("Top 10 Customers by Transaction Value")
plt.xlabel("Customer ID")
plt.ylabel("Total Transaction Value (USD)")
plt.show()

top_products = transactions_df.groupby("ProductID")["Quantity"].sum().nlargest(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_products.index, y=top_products.values, palette="viridis")
plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Product ID")
plt.ylabel("Total Quantity Sold")
plt.show()

product_revenue = transactions_df.groupby("ProductID")["TotalValue"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=product_revenue.head(10).index, y=product_revenue.head(10).values, palette="coolwarm")
plt.title("Top 10 Products by Revenue Contribution")
plt.xlabel("Product ID")
plt.ylabel("Total Revenue (USD)")
plt.show()

transactions_df["TransactionYearMonth"] = transactions_df["TransactionDate"].dt.to_period("M")
avg_transaction_value = transactions_df.groupby("TransactionYearMonth")["TotalValue"].mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x=avg_transaction_value.index.astype(str), y=avg_transaction_value.values, marker="o")
plt.title("Average Transaction Value Over Time")
plt.xlabel("Year-Month")
plt.ylabel("Average Transaction Value (USD)")
plt.xticks(rotation=45)
plt.show()
```
